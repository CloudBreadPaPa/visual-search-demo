## Step 1: fingerprinting the dataset

### Import the necessary packages

```python
from PIL import Image
import imagehash
#import argparse
#import shelve
import glob
import os
#import sys
#import numpy as np
#import pandas as pd
```

### Params

```python
dataset = os.getcwd()
#shelvefile  = os.path.join(os.getcwd(), "dups", "shelvedb")
pathpattern_ds = os.path.join(dataset,"dups","data","*","*.jpg")
#shelvefile
dataset
```

### open the shelve database


```python
#db = shelve.open(shelvefile, writeback = True)
```

### loop over the image dataset

```python
%%time
imgdict = {}
for imagePath in glob.glob(pathpattern_ds):
    # load the image and compute the difference hash
    image = Image.open(imagePath)
    h = str(imagehash.dhash(image))

    # extract the filename from the path and update the database
    # using the hash as the key and the filename append to the
    # list of values
    filename = imagePath[imagePath.rfind("/") + 1:]
#    db[h] = db.get(h, []) + [filename]
#    imgdict[h]= filename
    imgdict[h]= imagePath

# close the shelf database
#db.close()
```

### loop over the images and compute hamming distances for query image

```python
%%time
from operator import itemgetter
#import numpy as np
#n=len(imgdict.keys())
topn = 5
dist=[]
# loop over the images
#for imghash in imgdict.keys():
#    #img=mpimg.imread(filename)
#    #imgplot = plt.imshow(img)
#    #print(filename)
#    dist[imghash]=dis.hamming(imghash, h)
i=0
for imh1 in imgdict.keys():
#    j=0
    tdict={}
    for imh2 in imgdict.keys():
        tdict[imgdict[imh2]]=dis.hamming(imh1, imh2)
#        j=j+1
#    tdict= sorted(tdict, key=tdict.get, reverse=True)
    t=sorted(tdict.items(), key=itemgetter(1))
    dist[i:]=[t[0:topn]]
    i=i+1
print(dist[0])
```
