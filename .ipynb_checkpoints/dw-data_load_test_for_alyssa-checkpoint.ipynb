# Too slow, loading CSV data to np array

```python
import numpy as np
from PIL import Image
from os import listdir
import os
from os.path import isfile, join, splitext
import shutil
import random
import csv
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
# set real vector data
folder = 'pg_data/512d_embedding'
image_folder = 'pg_data/images'
files = [file for file in os.listdir(folder) if os.path.isfile(join(folder, file))]
```

```python
len(files)
```

```python
files[0:10]
```

```python
def flatten_2d(your_list):
    return [x for sublist in your_list for x in sublist]
```

```python
# change np array to list method
import time
init_time = time.time()
start = time.time()

all_list = []
i = 1
for f in files:
    with open(join(folder, f), 'r') as fi:
        reader = csv.reader(fi)
        filename = splitext(f)[0]
        #print('filename : ', filename)
        list1 = flatten_2d(list(reader))
        if i % 100 == 0:
            end = time.time()
            time_taken = end - start
            print('current process :', i, '\nTime taken : ', time_taken)
        list1.append(filename)
        #print('list1 : ', list1)
        all_list.append(list1)
        #print('all_list : ', all_list)
        i = i + 1

# load to np
arr_with_filename = np.array(all_list)
print('arr_with_filename.shape : ', arr_with_filename.shape)
print('Total time taken : ', time.time() - init_time)
```

```python
# create work array without filename
arr = arr_with_filename[:, :512]
arr.shape
```

```python
xb4 = arr.astype('float32')
```

```python
xq4 = xb4[102:103]  # set query vector
```

```python
print('xb4.shape : ',xb4.shape)
print('xq4.shape', xq4.shape)
```

```python
import faiss                   # make faiss available

d = 512
index = faiss.IndexFlatL2(d)   # build the index
print('index.is_trained : ', index.is_trained)

index.add(xb4)                  # add vectors to the index
print('index.ntotal : ', index.ntotal)
```

```python
k = 5                          # we want to see 4 nearest neighbors
D, I = index.search(xq4, k)     # actual search
print('I[:5] : ', I[:k])                   # neighbors of the 5 first queries
print('D[:5] : ', D[:k])
```

```python
arr_result = np.empty([0,512])
i = 0
for i in range(k):
    print('index : ', I[0,i], '\ndistance : ', D[0,i], '\narr : ', xb4[I[0,i]])
    arr_result = np.vstack((arr_result, xb4[I[0,i]]))
arr_result
```

```python
def get_map_image_filename(idx):
    #print(arr_with_filename[idx, -1])
    image_filename = arr_with_filename[idx, -1]
    image_filename = image_filename.astype('str') + '.jpg'
    return image_filename
```

```python
def draw_vector_map_image5(I):
    plt.figure(figsize=(20,4))
    #print('I[0]', I[0])
    for index, (array_index) in enumerate(zip(I[0])):
        #print('array_index', array_index)
        #print('index : ', index)
        #print('array_index[index] : ', array_index[0])
        fname = get_map_image_filename(array_index[0])
        #print('fname : ', fname)
        im = Image.open(join(image_folder, fname))
        plt.subplot(1, 5, index + 1)
        plt.imshow(im)
        titlestr = 'i:' + str(array_index[0]) + '  filename:', fname
        plt.title(titlestr, fontsize = 10)
```

```python
draw_vector_map_image5(I)
```
